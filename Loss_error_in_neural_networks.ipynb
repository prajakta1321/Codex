```python
# using neural networks for ML to learn the relationship between X and Y

# importing the libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras

# defining and compiling the neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

model.compile(optimizer='sgd', loss='mean_squared_error')


# providing the data
xs = np.array([1.0, 3.0, 6.0, 9.0, 12.0, 15.0], dtype=float)
ys = np.array([-2.0, 0.0, 2.0, 4.0, 6.0, 8.0], dtype=float)

# training the neural network
model.fit(xs, ys, epochs=500)
```

```python
model.fit(xs, ys, epochs=50)
# it is observed that as the epoch increases the loss decreases
```

```python
print(model.predict([10.0]))

# the output should be Y = 3X + 1
```
